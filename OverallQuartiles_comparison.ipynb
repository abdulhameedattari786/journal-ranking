```python
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score

# Assuming you have a column named 'cluster_labels' containing cluster labels
cluster_labels = resultInnerNew2['Kmeanslabels'].values
resultInnerNew2['cluster_labels'] = resultInnerNew2['Kmeanslabels'].replace({0: 1, 1: 3, 2: 0, 3: 2})
# Assuming you have columns for IF_Quartiles, SJR_Best_Quartiles, and Scopus_Quartiles
if_quartiles = resultInnerNew2['IF_Quartile'].values
sjr_quartiles = resultInnerNew2['SJR Best Quartile'].values
scopus_quartiles = resultInnerNew2['Scopus_Quartile'].values

# Define mapping for quartiles
quartile_mapping_SJR = {'Q1': 0, 'Q2': 1, 'Q3': 2, 'Q4': 3}
quartile_mapping_IF = {'IF_Q1': 0, 'IF_Q2': 1, 'IF_Q3': 2, 'IF_Q4': 3}
quartile_mapping_Scopus = {'PQ1': 0, 'PQ2': 1, 'PQ3': 2, 'PQ4': 3}
if_quartiles = pd.Series(if_quartiles)
sjr_quartiles = pd.Series(sjr_quartiles)
scopus_quartiles = pd.Series(scopus_quartiles)


# Map quartiles to numerical values
if_quartiles_numeric = if_quartiles.map(quartile_mapping_IF).values
sjr_quartiles_numeric = sjr_quartiles.map(quartile_mapping_SJR).values
scopus_quartiles_numeric = scopus_quartiles.map(quartile_mapping_Scopus).values
# Perform KMeans clustering (if necessary)
# Replace X with your feature matrix
# kmeans = KMeans(n_clusters=num_clusters)
# cluster_labels = kmeans.fit_predict(X)
ari_SJR = adjusted_rand_score(cluster_labels, sjr_quartiles_numeric)
ari_if = adjusted_rand_score(cluster_labels, if_quartiles_numeric)
ari_Scopus = adjusted_rand_score(cluster_labels, scopus_quartiles_numeric)
# mi_if = adjusted_mutual_info_score(cluster_labels, if_quartiles_numeric)
print(ari_SJR)
print(ari_if)
print(ari_Scopus)
```
